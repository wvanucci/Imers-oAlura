```python
df_acoes = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Total_de_acoes")
#df_acoes.head(10)
df_acoes
```

```python
df_ticker = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Ticker")
#df_acoes.head(10)
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_perc'}).copy()
df_principal
```

```python
df_principal['var_pct']=df_principal['var_dia_perc']/100
df_principal['valor_inicial']=df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal
```

**Juntar colunas de tabelas diferentes (analogo a PROCV no excel)
método .merge()**:

 .merge(*tabela_quer_quer_integrar_na_principal*,left_on='referencia tabela principal', right on ='referencia tabela secundaria', how='left')

Explicando:


*   left_on é o ccomando para dizer o critério (nome da coluna) da tabela principal
*   right_on é o comando para informar o critério da principal mas que está a tabela secundária
*   how= é o comando para integrar na tabela left (que é a nossa principal)
*   importante: os nomes das colunas devem ser iguais nas aspas ''

OBS: Caso nao tenha a mesma quantidad e de iterável na tabela secundária, vair retornar o valor null(vazio?)






```python
# Conectando data frames
df_principal=df_principal.merge(df_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
# tirando coluna de df
# acho que é intuitivo
df_principal=df_principal.drop(columns=['Código'])
df_principal
```

```python
# coluna de variação no dia = (vvalor_final - valor_inicial) * qrtde teórica
df_principal['variacao'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']

df_principal
```

```python
# formatar todos os valores do tipo float do notebook (utilizando pandas)
# Note: Eu acesso as opções do pandas (options), exibiçaõ de numeros float irei formatar (display.float_format)
# :.2f duas casas decimais
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
# Formatação dos iteraveis de coluna específica
# Explicando: coluna recebe ela mesma com o tipo desejado (astype(int))
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
# verificando tipos
tipos = df_principal['variacao']
df_principal.info()
```

-criar uma coluna resultados. Coluan se subiu ou desceu
-Utilizando chatGpt, prompts:


---


=SE(O2>0;"Subiu";SE(O2<0;"Desceu";"Estável"))
tenho uma planilha excel que utilizo essa fórmula e queria aplicar a mesma operação em um dataframe do python em uma linha. Utilize pandas e lambda


---


Eu quero comparar se valores de uma coluna de DataFrame está acima de zero, abaixo ou se é igual a zero. Essa comparação deve ser armazenada em outra coluna do mesmo DataFrame. Qual o código utilizando pandas e lambda


---



```python
# A função apply é utilizada para aplicar a condição (utilizando função anonima lambda) a cada valor da coluna "Coluna a comparar".

df_principal['resultado'] = df_principal['variacao'].apply(lambda x: 'Subiu' if x>0 else ('Estável' if x == 0 else 'Desceu' ) )
# criei nova coluna e vai receber a comparação da coluna varicao (utilizando lambda)
# existe a função .assign que é mais seguro pois faz uma cópia do dataframe (não modifica ela)
df_principal
```

```python
# 'procv' com nome de empresa
df_principal=df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left').drop(columns=['Ticker'])
df_principal
```

```python
# integrando colunas segmento e idade
df_principal=df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left').drop(columns=['Nome da Empresa'])
df_principal
```

```python
# Condiçao de categoria de idade
df_principal['cat_idade'] = df_principal['Idade (em anos)'].apply(lambda x: 'Mais de 100' if x>100 else ('Menos que 50' if x < 50 else 'Entre 50 e 100') )
df_principal
```

**Fazer analises (métricas)**

Procurar no chatGPT ou usar os paramêtros min(), max(), mean() direto e formatar no print

```python
media_var= df_principal['variacao'].mean()
max_var=df_principal['variacao'].max()
minx_var=df_principal['variacao'].min()

#filtragem por linha
med_subiu = df_principal[df_principal['resultado']=='Subiu']['variacao'].mean()
med_desceu = df_principal[df_principal['resultado']=='Desceu']['variacao'].mean()
med_estavel = df_principal[df_principal['resultado']=='Estável']['variacao'].mean()

print(f'Média da variação diária: R${media_var:,.2f}\n\nValor máximo do dia: R${max_var:,.2f}\n\nValor mínimo do dia: R${minx_var:,.2f}\n\n',end='')
print(f'Média diária quem subiu: R${med_subiu:,.2f}\n\nMédia diária quem ficou estável: R${med_estavel:,.2f}\n\nMédia diária quem desceu: R${med_desceu:,.2f}\n\n')
```

```python
# analise por segumento. Criando DF com os que subiram
df_principal_subiu = df_principal[df_principal['resultado']=='Subiu']
df_principal_subiu
```

```python
# groupby cria um objeto de agrupamento de dados unicos e faz a soma (sum()) correspondente a coluna especifica (varicao)
# Soma da vairaçao diária de cada tipo de segmento
# reset_index: convertre esse objeto em um dataframe
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao'].sum().reset_index()
df_analise_segmento
```

```python
# Analise do saldo (soma) das variações de quem subiu e desceu
df_analise_saldo = df_principal.groupby('resultado')['variacao'].sum().reset_index()
df_analise_saldo
```

**Gráficos**

utilizar a biblioteca plotly.express ( chamei de px as px)

```python
# Analise de saldo
# criando a figura
# fazer grafico do dataframe de analise saldo
fig = px.bar(df_analise_saldo, x='resultado', y='variacao', text='variacao', title='Reais por Resultado')
fig.update_traces(textposition="outside",texttemplate='R$%{text:,.2f}')
fig.update_layout(yaxis_title='Variação',xaxis_title='Faixa etária',yaxis_tickprefix='R$', yaxis_tickformat=',.2f')
#textposition="outside" deixar no topo do gráfico e  texttemplate='R$%{text:,.2f}' formatar o texto (%) marcado, no caso o text
# text adciona a infromaçao que vc quer na barra
fig.show()
```

```python
# Gráfico de segmentos

fig_pizza = px.pie(df_analise_segmento, values='variacao', names='Segmento', title='Saldo de quem subiu para cada segmento' )
fig.update_traces(textposition='inside')

fig_pizza.show()
```
